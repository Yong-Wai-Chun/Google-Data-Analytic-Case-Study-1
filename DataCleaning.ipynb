```python
import numpy as py
import pandas as pd
```

```python
df = pd.read_csv('data1/202108-divvy-tripdata.csv')
df1 = pd.read_csv('data1/202109-divvy-tripdata.csv')
df2 = pd.read_csv('data1/202110-divvy-tripdata.csv')
df3 = pd.read_csv('data1/202111-divvy-tripdata.csv')
df4 = pd.read_csv('data1/202112-divvy-tripdata.csv')
df5 = pd.read_csv('data1/202201-divvy-tripdata.csv')
df6 = pd.read_csv('data1/202202-divvy-tripdata.csv')
df7 = pd.read_csv('data1/202203-divvy-tripdata.csv')
df8 = pd.read_csv('data1/202204-divvy-tripdata.csv')
df9 = pd.read_csv('data1/202205-divvy-tripdata.csv')
df10 = pd.read_csv('data1/202206-divvy-tripdata.csv')
df11 = pd.read_csv('data1/202207-divvy-tripdata.csv')
```

```python
def cleaningFunc(df_para):
    df_para['started_time'] = df_para['started_at'].str[11:]
    df_para['ended_time'] = df_para['ended_at'].str[11:]
    
    df_para['started_at'] = df_para['started_at'].str[:10]
    df_para['ended_at'] = df_para['ended_at'].str[:10]
    
    df_para.sort_values(by=['started_at'], inplace=True)
    df_para.drop_duplicates(inplace=True)
    
    df_new1 = df_para[df_para['started_at'] < df_para['ended_at']]
    
    return df_new1
```

```python
df_new = cleaningFunc(df)
df_new1 = cleaningFunc(df1)
df_new2 = cleaningFunc(df2)
df_new3 = cleaningFunc(df3)
df_new4 = cleaningFunc(df4)
df_new5 = cleaningFunc(df5)
df_new6 = cleaningFunc(df6)
df_new7 = cleaningFunc(df7)
df_new8 = cleaningFunc(df8)
df_new9 = cleaningFunc(df9)
df_new10 = cleaningFunc(df10)
df_new11 = cleaningFunc(df11)
```

```python
df_members = df_new[df_new['member_casual'] == "member"]
df_members1 = df_new1[df_new1['member_casual'] == "member"]
df_members2 = df_new2[df_new2['member_casual'] == "member"]
df_members3 = df_new3[df_new3['member_casual'] == "member"]
df_members4 = df_new4[df_new4['member_casual'] == "member"]
df_members5 = df_new5[df_new5['member_casual'] == "member"]
df_members6 = df_new6[df_new6['member_casual'] == "member"]
df_members7 = df_new7[df_new7['member_casual'] == "member"]
df_members8 = df_new8[df_new8['member_casual'] == "member"]
df_members9 = df_new9[df_new9['member_casual'] == "member"]
df_members10 = df_new10[df_new10['member_casual'] == "member"]
df_members11 = df_new11[df_new11['member_casual'] == "member"]
```

```python
df_casuals = df_new[df_new['member_casual'] == "casual"]
df_casuals1 = df_new1[df_new1['member_casual'] == "casual"]
df_casuals2 = df_new2[df_new2['member_casual'] == "casual"]
df_casuals3 = df_new3[df_new3['member_casual'] == "casual"]
df_casuals4 = df_new4[df_new4['member_casual'] == "casual"]
df_casuals5 = df_new5[df_new5['member_casual'] == "casual"]
df_casuals6 = df_new6[df_new6['member_casual'] == "casual"]
df_casuals7 = df_new7[df_new7['member_casual'] == "casual"]
df_casuals8 = df_new8[df_new8['member_casual'] == "casual"]
df_casuals9 = df_new9[df_new9['member_casual'] == "casual"]
df_casuals10 = df_new10[df_new10['member_casual'] == "casual"]
df_casuals11 = df_new11[df_new11['member_casual'] == "casual"]
```

```python
df_members_list = [df_members,df_members1,df_members2,df_members3,
                   df_members4,df_members5,df_members6,df_members7,
                   df_members8,df_members9,df_members10,df_members11,]
df_casuals_list = [df_casuals,df_casuals1,df_casuals2,df_casuals3,
                   df_casuals4,df_casuals5,df_casuals6,df_casuals7,
                   df_casuals8,df_casuals9,df_casuals10,df_casuals11,]
```

```python
df_members_concat = pd.concat(df_members_list)
df_casuals_concat = pd.concat(df_casuals_list)
```
